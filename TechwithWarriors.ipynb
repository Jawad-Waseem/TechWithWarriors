# Task 01


**Perform basic data cleaning and exploratory data
analysis using pandas.**

Use pandas to clean and explore a
dataset, performing operations like handling missing
values and generating summary statistics.

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/Titanic.csv')

# Handle missing values - Select only numerical columns

numeric_columns = df.select_dtypes(include=['number'])
numeric_columns.fillna(numeric_columns.mean(), inplace=True)
df[numeric_columns.columns] = numeric_columns

# Generate summary statistics

print(df.describe())

# Explore data distribution

print(df['Survived'].value_counts())  # Count of survivors and non-survivors
```

# Task 02

**Create visualizations such as bar charts and
scatter plots using Matplotlib or Seaborn.**

 Generate visualizations to represent
data distributions and relationships using Matplotlib
or Seaborn.

```python


import matplotlib.pyplot as plt
import seaborn as sns

# Example: Bar chart of passenger survival by class

sns.barplot(x='Pclass', y='Survived', data=df)
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Passenger Class')
plt.show()

# Example: Scatter plot of age vs. fare

plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Age vs. Fare')
plt.show()
```

# Task 03

**Implement a time series analysis to forecast sales
data using ARIMA or Prophet.**

Use ARIMA or Prophet to analyze and
forecast sales data over time.

```python
# prompt: Use ARIMA or Prophet to analyze and forecast sales data over time do this and according to this requirment use faker to make a dataset accordingly and solve this

import matplotlib.pyplot as plt
import pandas as pd
from prophet import Prophet

# Initialize Faker
fake = Faker()

# Generate fake sales data
data = []
start_date = pd.to_datetime('2022-01-01')
for _ in range(365):
  date = start_date
  sales = fake.random_int(min=50, max=200)
  data.append({'ds': date, 'y': sales})
  start_date += pd.Timedelta(days=1)

df = pd.DataFrame(data)

# Create and fit the Prophet model


model = Prophet()
model.fit(df)

# Forecast future sales


future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Plot the forecast

fig = model.plot(forecast)
plt.show()
```
